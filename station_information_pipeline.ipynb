```python
def create_plotly_map(df):
    """
    Create interactive map with Plotly
    - Works great in Jupyter notebooks
    - Hover for information
    - Color-coded by values
    """
    fig = px.scatter_mapbox(
        df,
        lat='lat',
        lon='lon',
        hover_name='name',
        hover_data=['capacity'],
        color='capacity',
        size='capacity',
        color_continuous_scale='Viridis',
        zoom=12,
        height=600,
        title='Bike Stations Map'
    )
    
    fig.update_layout(
        mapbox_style='open-street-map',
        margin={"r": 0, "t": 40, "l": 0, "b": 0}
    )
    
    #fig.show()
    return fig
```

```python
import plotly.express as px
create_plotly_map(df_map)
```
