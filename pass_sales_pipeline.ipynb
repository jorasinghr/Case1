```python
import utils.functions as fnc
base_api_url='https://oslo-bike-api-13322556367.europe-west1.run.app'
```

```python
import pandas as pd
pass_sales = fnc.LocalStorage()

pass_sales_df = pass_sales.get_data(base_api_url+'/pass_sales', ['data'])
```

```python
pass_sales_df
```

```python
bronze_df = pass_sales.load_to_bronze(pass_sales_df, 'pass_sales')
```

```python
bronze_df
```

```python
silver_df = pass_sales.transform_to_silver(bronze_df, 'pass_sales',
                                     transformations = pass_sales.transform_stations,
                                     quality_checks = pass_sales.quality_checks_2)
```

```python
silver_df
```

```python
def aggregation_3(df):
    return {
        'pass_type_count': df.groupby('pass_type')['transaction_id'].count().reset_index(),
        'pass_type_sum': df.groupby('pass_type')['price_nok'].sum().reset_index()

    }
```

```python
gold_df = pass_sales.aggregate_to_gold(silver_df, 'pass_sales', aggregations=aggregation_3)
```
